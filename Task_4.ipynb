```python
import gradio as gr
import datetime
import pytz
import torch
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from transformers import pipeline
```

```python
# English Speech Recognition Model
speech_to_text = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-small",
    device=0 if torch.cuda.is_available() else -1
)

# English → Hindi Translation Model
en_hi_translator = pipeline(
    "translation",
    model="Helsinki-NLP/opus-mt-en-hi"
)
```

```python
def is_active_time():
    ist = pytz.timezone("Asia/Kolkata")
    now = datetime.datetime.now(ist).time()

    start = datetime.time(21, 30)  # 9:30 PM
    end = datetime.time(22, 0)     # 10:00 PM

    return start <= now <= end
```

```python
def voice_translate(audio):
    # Time check
    if not is_active_time():
        return "Taking rest, see you tomorrow!"

    if audio is None:
        return "Please repeat it for better accuracy."

    try:
        # Speech to text
        result = speech_to_text(audio)
        english_text = result["text"].strip()

        # If unclear audio
        if len(english_text) < 3:
            return "Please repeat it for better accuracy."

        # Translate to Hindi
        hindi = en_hi_translator(english_text)[0]["translation_text"]

        return f"""
English Text:
{english_text}

Hindi Translation:
{hindi}
"""

    except Exception:
        return "Please repeat it for better accuracy."
```

```python
interface = gr.Interface(
    fn=voice_translate,
    inputs=gr.Audio(type="filepath", label="Speak English"),
    outputs=gr.Textbox(label="Hindi Output"),
    title="English to Hindi Voice Translator",
    description="""
• Works ONLY between 9:30 PM – 10:00 PM (IST)
• Input must be ENGLISH audio
• If audio is unclear → asks to repeat
"""
)
```

```python
if __name__ == "__main__":
    interface.launch(share=True)
```
