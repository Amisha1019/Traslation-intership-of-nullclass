```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import gradio as gr
import datetime
import pytz
from transformers import pipeline
```

```python
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-hi")
```

```python
def is_vowel_allowed_time():
    ist = pytz.timezone("Asia/Kolkata")
    now = datetime.datetime.now(ist).time()

    start = datetime.time(21, 0)  # 9:00 PM
    end = datetime.time(22, 0)    # 10:00 PM

    return start <= now <= end
```

```python
def translate_english_to_hindi(word):
    if not word:
        return "Please enter a word."

    word = word.strip()
    first_char = word[0].lower()

    vowels = ['a', 'e', 'i', 'o', 'u']

    # If word starts with vowel
    if first_char in vowels:
        if not is_vowel_allowed_time():
            return "This word starts with a vowel. Please provide another word."
        # else: allowed during 9–10 PM

    # Translate
    hindi = translator(word)[0]['translation_text']
    return f"Hindi Translation: {hindi}"
```

```python
interface = gr.Interface(
    fn=translate_english_to_hindi,
    inputs=gr.Textbox(
        label="Enter English Word",
        placeholder="example: apple / computer"
    ),
    outputs=gr.Textbox(label="Output"),
    title="Logic-Based English → Hindi Translator",
    description="""
Rules:
• Words starting with vowels are NOT translated
• EXCEPTION: Allowed between 9 PM – 10 PM
• Consonant-starting words are always translated
"""
)
```

```python
if __name__ == "__main__":
    interface.launch(share=True)
```
