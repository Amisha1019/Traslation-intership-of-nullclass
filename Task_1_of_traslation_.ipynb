```python
from transformers import pipeline
import gradio as gr

# French → English (public)
fr_en = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")

# English → Tamil (public)
en_ta = pipeline("translation", model="Helsinki-NLP/opus-mt-en-ta")
```

```python
def translate_french_to_tamil(word):
    word = word.strip()

    if len(word) != 5:
        return "❌ Only French words with EXACTLY 5 letters are allowed."

    # French → English
    english = fr_en(word)[0]['translation_text']

    # English → Tamil
    tamil = en_ta(english)[0]['translation_text']

    return f"""
French Word : {word}
English     : {english}
Tamil       : {tamil}
"""
```

```python
interface = gr.Interface(
    fn=translate_french_to_tamil,
    inputs=gr.Textbox(label="J'aime le thé.", placeholder="merci"),
    outputs=gr.Textbox(label="Translation Output"),
    title="French ➜ Tamil Translation App (5-Letter Rule)",
    description="Uses Machine Learning. Translation works ONLY for exactly 5-letter French words."
)
```

```python
interface.launch(share=True)
```
