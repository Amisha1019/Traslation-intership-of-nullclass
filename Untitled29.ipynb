```python
from transformers import pipeline
import gradio as gr

# Translation pipelines (public models)
en_fr = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr")
en_hi = pipeline("translation", model="Helsinki-NLP/opus-mt-en-hi")

def dual_translate(text):
    text = text.strip()

    if len(text) < 10:
        return "upload again", "upload again"

    french = en_fr(text)[0]["translation_text"]
    hindi = en_hi(text)[0]["translation_text"]

    return french, hindi

interface = gr.Interface(
    fn=dual_translate,
    inputs=gr.Textbox(label="Enter English text (10+ letters)"),
    outputs=[
        gr.Textbox(label="French Translation"),
        gr.Textbox(label="Hindi Translation")
    ],
    title="Dual Language Translator",
    description="Translates English to French and Hindi if text has 10 or more letters."
)

if __name__ == "__main__":
    interface.launch()
```
